# 02. Model Training and Evaluation

This notebook demonstrates step-by-step preprocessing, training, and evaluation of classification models for network intrusion detection. We will load the preprocessed data, evaluate Decision Trees and Random Forests, and analyze their confusion matrices.

```python
import os
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Import custom src modules
import sys
sys.path.append('..')
from src.config import BINARY_TARGET, MULTICLASS_TARGET

%matplotlib inline
sns.set_theme(style='whitegrid')
```

## 1. Load Preprocessed Datasets

```python
train_df = pd.read_csv('../data/train_processed.csv')
test_df = pd.read_csv('../data/test_processed.csv')

target_columns = [BINARY_TARGET, MULTICLASS_TARGET]
feature_columns = [col for col in train_df.columns if col not in target_columns]

X_train, y_train_bin = train_df[feature_columns], train_df[BINARY_TARGET]
X_test, y_test_bin = test_df[feature_columns], test_df[BINARY_TARGET]

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
```

## 2. Load and Test the Binary Classifier

Let's load the trained Binary Random Forest model.

```python
bin_rf = joblib.load('../models/binary_rf.joblib')
y_pred = bin_rf.predict(X_test)

print(f"Accuracy Score: {accuracy_score(y_test_bin, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test_bin, y_pred, target_names=['Normal', 'Anomaly']))
```

## 3. Visualize Confusion Matrix

```python
cm = confusion_matrix(y_test_bin, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'Anomaly'], yticklabels=['Normal', 'Anomaly'])
plt.title('Binary Random Forest Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
