# 01. Exploratory Data Analysis (EDA) - NSL-KDD Dataset

This notebook demonstrates exploratory data analysis of the NSL-KDD intrusion detection dataset. We will explore feature distributions, check for missing values, analyze target class frequencies, and visualize relationships between protocol types and attack categories.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import configuration and loader from src
import sys
sys.path.append('..')
from src.config import RAW_TRAIN_PATH, COLUMNS, ATTACK_MAPPING
from src.data_loader import get_datasets, load_raw_data

%matplotlib inline
sns.set_theme(style='whitegrid')
```

## 1. Load Dataset

Let's load the training dataset using the data loader we built.

```python
train_df = load_raw_data(RAW_TRAIN_PATH)
print(f"Dataset shape: {train_df.shape}")
train_df.head()
```

## 2. Basic Dataset Statistics

```python
train_df.describe().T
```

## 3. Analyze Target Variable (Attack Types)

Let's map individual attack types to their broader attack classes (DoS, Probe, R2L, U2R) and analyze their distribution.

```python
# Binary label check
train_df['label_binary'] = train_df['attack_type'].apply(lambda x: 'Normal' if x == 'normal' else 'Anomaly')

plt.figure(figsize=(6, 4))
sns.countplot(x='label_binary', data=train_df, palette='viridis')
plt.title('Binary Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Multiclass label mapping
train_df['label_multiclass'] = train_df['attack_type'].apply(lambda x: ATTACK_MAPPING.get(x, 'dos'))

plt.figure(figsize=(8, 5))
sns.countplot(x='label_multiclass', data=train_df, order=train_df['label_multiclass'].value_counts().index, palette='muted')
plt.title('Multiclass Attack Categories Distribution')
plt.xlabel('Attack Class')
plt.ylabel('Count')
plt.show()
```

## 4. Explore Categorical Features

Let's visualize the relationship between categorical variables like `protocol_type` and target classes.

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='protocol_type', hue='label_binary', data=train_df, palette='Set2')
plt.title('Protocol Type by Binary Target')
plt.xlabel('Protocol Type')
plt.ylabel('Count')
plt.legend(title='Label')
plt.show()
```
